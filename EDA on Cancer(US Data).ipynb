# D'Agostino & Pearson's test to identify normally distributed data & outlier Detection

```python
processed_data = pd.read_csv(r"D:\datasets\cancer_reg_processed.csv")
processed_data
```

```python
from scipy.stats import normaltest

numerical_cols = processed_data.select_dtypes(np.number).columns
gaussian_cols=[]
nonGausian_cols=[]

for col in numerical_cols:
    testStat, Pval = normaltest(processed_data[col])
    
    alpha = 0.05
    if Pval < alpha:
        nonGausian_cols.append(col)
    else:
        gaussian_cols.append(col)
        
print(gaussian_cols)
        
        
```

```python
#plot all the Gaussian CCols

fig, axes =plt.subplots(nrows=1, ncols=1, figsize=(10,10))
  
for i, col in enumerate(gaussian_cols):
    processed_data[col].plot(kind = 'density', ax=axes, subplots=True, sharex=False)
plt.tight_layout()
plt.show()
```

```python
gaussian_data =  processed_data[gaussian_cols]
basic_gaussian_info = gaussian_data.describe()
basic_gaussian_info
```

```python
#outliers removal


highest_allowed =  basic_gaussian_info['pctpubliccoverage']['mean'] + 3*basic_gaussian_info['pctpubliccoverage']['std']
lower_allowed   =  basic_gaussian_info['pctpubliccoverage']['mean'] - 3*basic_gaussian_info['pctpubliccoverage']['std']
print('highest_allowed  = ', highest_allowed, ' lower_allowed = ', lower_allowed)

processed_data[(processed_data['pctpubliccoverage'] < lower_allowed) | (processed_data['pctpubliccoverage'] > highest_allowed)]
```

```python
#function for detecting outliers from all columns

def get_outliers(df, col, basic_info):
    highest_allowed =  basic_gaussian_info[col]['mean'] + 3*basic_gaussian_info['pctpubliccoverage']['std']
    lower_allowed   =  basic_gaussian_info[col]['mean'] - 3*basic_gaussian_info['pctpubliccoverage']['std']
    
    outlier_df = processed_data[(processed_data['pctpubliccoverage'] < lower_allowed) | (processed_data['pctpubliccoverage'] > highest_allowed)]
    return outlier_df

col_have_outliers = []

for col in gaussian_cols:
    df = get_outliers(gaussian_data, col, basic_gaussian_info)
    if df.shape[0]>0:
        col_have_outliers.append(col)
        
print(col_have_outliers)
      
```

```python
#trimming

for cols in col_have_outliers:
    highest_allowed =  basic_gaussian_info[col]['mean'] + 3*basic_gaussian_info['pctpubliccoverage']['std']
    lower_allowed   =  basic_gaussian_info[col]['mean'] - 3*basic_gaussian_info['pctpubliccoverage']['std']
    
    trimmed_data = processed_data[(processed_data['pctpubliccoverage'] > lower_allowed) & (processed_data['pctpubliccoverage'] < highest_allowed)]

trimmed_data
    
```

```python
#capping

for cols in col_have_outliers:
    highest_allowed =  basic_gaussian_info[col]['mean'] + 3*basic_gaussian_info['pctpubliccoverage']['std']
    lower_allowed   =  basic_gaussian_info[col]['mean'] - 3*basic_gaussian_info['pctpubliccoverage']['std']
    
    capped_data = processed_data.copy()
    capped_data.loc[capped_data[col]>highest_allowed, col] = highest_allowed
    capped_data.loc[capped_data[col]<lower_allowed,   col] = lower_allowed

capped_data
```

#To identify skewness

```python
#remove columns which are categorical & have less than 10 uniwue columns

cols_to_remove = []
for cols in processed_data.columns:
    if processed_data[cols].nunique() < 10:
        cols_to_remove.append(cols)
        
print(len(cols_to_remove))
data_for_skewness = processed_data.drop(cols_to_remove, axis=1)
data_for_skewness
```

```python


def identify_skewed_cols(df, cols):
    skewed_cols = []
    
    for col in cols:
        skew = df[col].skew()
        if skew<-1 or skew>1:
            skewed_cols.append(col)
    
    return skewed_cols
            
            
skewed_cols =  identify_skewed_cols(data_for_skewness, data_for_skewness.columns)
len(skewed_cols)
```

```python
skewed_data = data_for_skewness[skewed_cols]
skewed_data
```

```python
# plot the distribution of skewed columns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20,10))

for i,col in enumerate(skewed_cols):
    ax[i//4, i%4].set_title(col)
    skewed_data[col].plot(kind='density', ax=axes[i//4, i%4], subplots=True, sharex=False)
    
plt.tight_layout()
plt.show()
    
```

```python

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20,10))

for i,col in enumerate(skewed_cols):
    ax[i//4, i%4].set_title(col)
    skewed_data[col].plot(kind='box', ax=axes[i//4, i%4], subplots=True, sharex=False)
    
plt.tight_layout()
plt.show()
```

```python
def capping(df, col):
    percentile75 = df[col].quantile(0.75)
    percentile25 = df[col].quantile(0.25)
    
    iqr =  percentile75 - percentile25
    upper_limit =  percentile75 + 1.5*iqr 
    lower_limit =  percentile25 - 1.5*iqr
    
    df.loc[df[col]>upper_limit, col] = upper_limit
    df.loc[df[col]<lower_limit, col] = lower_limit
    
    return df

    
for col in skewed_cols:
    capped_data = capping(processed_data, col)
capped_data
    
```

```python
#box plot of capped data

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20,10))

for i,col in enumerate(skewed_cols):
    ax[i//4, i%4].set_title(col)
    capped_data[col].plot(kind='box', ax=axes[i//4, i%4], subplots=True, sharex=False)
    
plt.tight_layout()
plt.show()
```
